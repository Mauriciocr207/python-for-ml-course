# Introducción a NumPy

**NumPy** (Numerical Python) es una de las bibliotecas fundamentales del ecosistema científico de Python.  
Proporciona herramientas poderosas para trabajar con **arreglos multidimensionales** y realizar cálculos numéricos de manera **rápida y eficiente**.

## ¿Por qué usar NumPy?

- **Velocidad**: Las operaciones se ejecutan en código optimizado en C, mucho más rápido que las listas nativas de Python.  
- **Arreglos (ndarray)**: Permite trabajar con vectores y matrices de forma intuitiva.  
- **Funciones matemáticas avanzadas**: Incluye álgebra lineal, transformadas de Fourier, estadísticas, entre otras.  
- **Compatibilidad**: Se integra fácilmente con bibliotecas como *Pandas*, *Matplotlib*, *SciPy* y *scikit-learn*.

### Operaciones con arrays

Estadística

````python
normal_dist = np.random.normal(loc=0., scale=0.001, size=100000)
plt.hist(normal_dist, bins=1000)
plt.show()

print("media ", normal_dist.mean())
print("max ", normal_dist.max())
print("min ", normal_dist.min())
````

### Meshgrid

````python
a = np.linspace(1, 10, 10)
b = np.linspace(1, 10, 10)

A, B = np.meshgrid(a, b)
print(A)
print(B)

values = A + B

plt.imshow(values, cmap='gray')    # 'gray' para escala de grises
plt.axis('off')
plt.show()
````

### Generemos una imagen con ruido gris

````python
img = np.random.rand(256, 256)  # valores entre 0 y 1

plt.imshow(img, cmap='gray')    # 'gray' para escala de grises
plt.axis('off')
plt.show()
````

### Ahora generemos una imagen con distribución gausiana en el centro

````python
width, height = 100, 100

x = np.linspace(-1, 1, width)
y = np.linspace(-1, 1, height)

X, Y = np.meshgrid(x, y)

sigma = 0.5

gaussian = np.exp(- (( X**2 + Y**2)) / (2 * sigma**2))

img = 1 - gaussian

plt.imshow(img, cmap="gray", origin="lower")
plt.axis("off")
plt.show()
````
